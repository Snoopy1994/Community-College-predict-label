```python
import numpy as np 
import pandas as pd
```

```python
with open("./keywords＿dic.txt", "r") as fd:
    data = fd.read()
data = data.splitlines()
data = [i.split(", ") for i in data]
```

```python
data
```

```python
dic = {}
for i, j in data:
    x = dic.get(j, set())
    x.add(i)
    dic[j] = x
```

```python
training_data = pd.read_csv("./110_training_0628.csv")
```

```python
training_data
```

```python
def get_labels(data, dictionary):
    results = []
    for i, js in dictionary.items():
        for j in js:
            if j not in data:
                continue
            results.append(i)
            break
    return results
```

```python
predict_labels_intro = []
for i in training_data["intro"]:
    label = get_labels(i, dic)
    predict_labels_intro.append(label)
```

```python
predict_labels_name = []
for i in training_data["name"]:
    label = get_labels(i, dic)
    predict_labels_name.append(label)
```

```python
predict_labels = []
for i, j in zip(predict_labels_intro, predict_labels_name):
    joined_labels = list(set(i+j))
    predict_labels.append(joined_labels)
```
